## Exploratory Data Analysis on Liver Disease Dataset
This dataset is related to liver function and disease detection from a medical diagnosis study. 

### Feature Description:
1. Age: Age of the patient.
2. Gender: Male or Female.
3. Total Bilirubin: A measure of bilirubin levels in the blood, which can indicate liver disease.
4. Direct Bilirubin: A specific form of bilirubin that helps diagnose liver conditions.
5. Alkaline Phosphatase: An enzyme related to liver and bone health.
6. Alamine Aminotransferase (ALT): An enzyme found in the liver; high levels may indicate liver damage.
7. Aspartate Aminotransferase (AST): Another enzyme that helps assess liver function.
8. Total Proteins: The total amount of proteins in the blood, indicating liver function.
9. Albumin: A major protein produced by the liver, essential for maintaining fluid balance.
10. Albumin and Globulin Ratio: The ratio of albumin to globulin, which helps in diagnosing liver diseases.
11. Dataset: This column appears to be a target variable (1 or 2), possibly indicating whether the patient has liver disease or is healthy.

## OUTLIER ANALYSIS
Outliers are data points that significantly deviate from the rest of the dataset. They can distort statistical analyses and affect machine learning models. Detecting and handling outliers is crucial for cleaning data and improving model accuracy.

```python
# boxplots are used to identify outlier in our data visually
```

```python
# columns wise boxplot
plt.boxplot(df['Total_Protiens'])
plt.show()
```

```python
# small bubble indicate the presence of outlier
# it is observed that there is presence of outlier in column 'Total_Protiens'
```

```python
# column wise boxplot for entire data
df.plot(kind='box', figsize=(20,10))
```

```python
# it is observed that outliers are present in many col
```

```python
# function is created for the identification of outliers
# function is designed to detect outliers in a given column using the Interquartile Range (IQR) method.
def identify_outliers(col_name):
    i=col_name
    # calculate quartiles
    Q1=round(np.quantile(df[i],0.25),2)
    Q2=round(np.quantile(df[i],0.50),2)
    Q3=round(np.quantile(df[i],0.75),2)
    # compute inter quartile range
    IQR=Q3-Q1
    # determine outlier bound
    upper_bound=Q3+(1.5*IQR)
    lower_bound=Q1-(1.5*IQR)
    # conditions to determine outlier
    cond1=df[i]>upper_bound
    cond2=df[i]<lower_bound
    # apply the conditions to find outlier
    outlier=df[i][cond1|cond2] #series
    # retrun dataframe with outlier
    outliers_data=outlier.values 
    return outliers_data
```

```python
outliers_data_chol=identify_outliers('Total_Protiens')
len(outliers_data_chol)
# no of outlier
```

```python
df.columns
```

```python
# number of outlier in each col
for i in df.columns:
    li=identify_outliers(i)
    print(i,len(li))
```

```python
# this function retruns the data frame having no outlier for that particular col
def non_outlier_df(col_name):
    i=col_name
    Q1=round(np.quantile(df[i],0.25),3)
    Q2=round(np.quantile(df[i],0.50),3)
    Q3=round(np.quantile(df[i],0.75),3)
    IQR=Q3-Q1
    upper_bound=Q3+(1.5*IQR)
    lower_bound=Q1-(1.5*IQR)
    cond1=df[i]<=upper_bound
    cond2=df[i]>=lower_bound
    # both cond need to be true
    non_outlier=df[cond1&cond2]
    return non_outlier
```

```python
# for the column name 'BloodPressure'
non_outlier_data=non_outlier_df('Total_Protiens')
```

```python
# this function retrun the plot for data containing outlier and data free from oulier
def plot_outlier_vs_non_outlier(col):
    i=col
    plt.figure(figsize=(12,8))
    plt.subplot(2,2,1)
    plt.title('non outlier')
    plt.hist(non_outlier_data[i],)
    
    plt.subplot(2,2,2)
    plt.title('outlier data')
    plt.hist(df[i])
    
    plt.subplot(2,2,3)
    plt.title('non outlier')
    plt.boxplot(non_outlier_data[i])
    
    plt.subplot(2,2,4)
    plt.title('outlier data')
    plt.boxplot(df[i])
    plt.show()
```

```python
# for column name 'Total_Protiens'
non_outlier_data=non_outlier_df('Total_Protiens')
plot_outlier_vs_non_outlier('Total_Protiens')
```

```python
# for column name 'Albumin_and_Globulin_Ratio'
non_outlier_data=non_outlier_df('Albumin_and_Globulin_Ratio')
plot_outlier_vs_non_outlier('Albumin_and_Globulin_Ratio')
```

```python
# function to remove the outliers passinf both dataframe and columns
# return the whole data frame free from outlier
def removeOutliers(df , cols):
    '''
    Removes outliers from multiple columns using the IQR method.
    '''
    # Create a copy to avoid modifying the original DataFrame
    df_cleaned = df.copy()
    for col in cols : 
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1 
        lower_bound = Q1-1.5*IQR
        upper_bound = Q3+1.5*IQR
        df_cleaned = df_cleaned[(df_cleaned[col]>=lower_bound) & (df_cleaned[col]<=upper_bound)]
    return df_cleaned
```

```python
# run function
liverData = removeOutliers(df,df.columns)
```

```python
liverData.plot(kind='box', figsize=(20,10))
```

```python
# number of outliers in data has been drastically reduced
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Training the Logistic Regression and model with Training data
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```
